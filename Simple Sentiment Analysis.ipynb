```python
from textblob import TextBlob
```

```python
from newspaper import Article
```

```python
url = 'https://en.wikipedia.org/wiki/Population_decline'
article = Article(url)
```

```python
import nltk
nltk.download('punkt')
```

```python
article.download()
article.parse()
article.nlp()
```

```python
text = article.summary
print(text)
```

```python
blob = TextBlob(text)
sentiment = blob.sentiment.polarity
```

```python
print(sentiment)
```
